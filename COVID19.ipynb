## Plot cases for countries by day on linear scale

Each country is centered on "day 0" -- they day they had at least 1500 cases.  This allows us to compare the 
rate of growth of the number of cases.  1500 was chosen to insure we are seeing national response -- before 
1500 cases, many countries are not reacting at the national level.  1500 cases also removes some minor variation that
we see in reporting of the numbers.

```r
# png(width=1100,height=1100,units="px",filename="cv19.png")
options(repr.plot.width=7, repr.plot.height=7) # Jupyter default is 7,7
matplot(cv19$X,cv19[,seq(2,11)],xlab="Days", ylab="Cases",type="o",lty=3)
legend(-9,35000,legend=c("1 - China","2 - Italy","3 - Iran","4 - South Korea","5 - Spain","6 - Germany", "7 - France", "8 - US", "9 - Switzerland", "0 - Netherlands"))
title("Coronavirus cases centered on Day 0 = 1500 cases")
# dev.off()
```

## Plot cases for countries by day on log scale

Each country is centered on "day 0" -- the day the country had at least 1500 cases.  This allows us to compare the 
rate of growth of the number of cases.  1500 was chosen to insure we are seeing national response -- before 
1500 cases, many countries are not reacting at the national level.  1500 cases also removes some minor variation that
we see in reporting of the numbers.

```r
# png(width=1100,height=1100,units="px",filename="cv19.png")
options(repr.plot.width=10, repr.plot.height=10) # Jupyter default is 7,7
countries<-c("China","Italy","US","Norway", "France","Spain","Netherlands", "South.Korea", 
             "Germany", "New.York", "Iran", "Switzerland") # countries to plot
cidx<-which(names(cv19)%in%countries)
matplot(cv19$X[seq(11,31)],cv19[seq(11,31),cidx],xlab="Days", 
        ylab="Cases",type="l",lty=1, log="y", lwd=3, 
        col=rainbow(length(countries)))
lines(seq(0,20),1.3^seq(0,20)*1500,lty=3)
lines(seq(0,20),1.2^seq(0,20)*1500,lty=3)
lines(seq(0,20),1.4^seq(0,20)*1500,lty=3)
lines(seq(0,20),1.5^seq(0,20)*1500,lty=3)
lines(seq(0,20),1.1^seq(0,20)*1500,lty=3)
legend(-0.5,80000,box.lty=0,fill=rainbow(length(countries)),legend=colnames(cv19)[cidx])
title("Coronavirus cases (log scale) Day 0 = 1500 cases")
# dev.off()
```
